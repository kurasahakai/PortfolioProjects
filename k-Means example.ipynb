# k-Means algorithm 
This is my take on the k-Means algorithm exploration proposed by jupyter nbviewer, it is to be intended as way for me to explore and learn more about k-Means.

The first task is to identify the positional average of groups of data. 

Let's first import scikit learn and generate some random blobs and plot them on a scatter plot

```python
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
#let's generate 4 random blobs of data on an X-Y plane
X, Y = datasets.make_blobs(centers=4, cluster_std=0.5, random_state=1234)
#Let's plot them on a scatter plot
plt.scatter(X[:,0], X[:,1]);
```

```python
#We can see the data has been created in 4 different groups by assigning to each group a color
plt.scatter(X[:,0], X[:,1], c=Y);
```

While we used the Y data to create the groups and assign a color to them, this is an artifact of how we created the dataset, however we usually have no information on Y. 
The point of k-Means is precisely to fin this information. 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4, random_state=8)
Y_hat = kmeans.fit(X).labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat);
```

We are, more often than not interested in groupings of this data or their positional means. We can find the means $\mu$ using the `cluster_centers_` method 

```python
import numpy as np
#let's plot the data with reduced opacity
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.3)
#find the coordinates of the cluster means
mu = kmeans.cluster_centers_
#let's plot a point for the position of the center with maximum opacity
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat))
print(mu)
```
